```python
#OR回路
import torch
inputData = torch.Tensor([[0, 0], [0, 1], [1, 0], [1, 1]])
outputData = torch.Tensor([[0], [1], [1], [1]])
```

```python
import torch.nn as nn

INPUT_NUM = 2
OUTPUT_NUM = 1

#活性化関数
activationFunc = nn.Sigmoid()

modelParam = nn.Sequential(
      nn.Linear(INPUT_NUM, OUTPUT_NUM),
      activationFunc,
)
```

```python
criterion = nn.MSELoss()  #損出関数定義定義
```

```python
optimizer = torch.optim.SGD(modelParam.parameters(),lr=0.1) #最適化のための計算定義
```

```python
modelParam.train()
for epoch in range(10000):
    optimizer.zero_grad()
    y=modelParam.forward(inputData.float())
    loss = criterion(y,outputData)
    loss.backward()
    optimizer.step()
    if epoch % 100 == 0:
        print(f"Epoch{epoch+1}", f"loss:{loss:.4f}")
```
