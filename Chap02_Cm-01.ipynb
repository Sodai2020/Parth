```python
#2
import math
import numpy as np
from numpy import random

N_in = 0
N_out = 0
N = 10000 #試行回数
x = random.rand(N) #Xの乱数
y = random.rand(N) #Yの乱数
EN = np.hypot(x,y) #X^2 + Y^2の平方根   #EN=ユークリッドのるむ

for i in EN:
    if i <= 1:
        N_in += 1
    else:
        N_out += 1

Pie = N_in/N*4 #パイの近似式

print("IN: {} ".format(N_in))
print("OUT: {} ".format(N_out))
print("ALL: {} ".format(N))

print("Pi: {} ".format(Pie))

#円の内部に入った数=IN(出力)
```

```python
#プロット
c1 = plt.Circle((0, 0), radius=1, fc="None", ec="r", linewidth=2, color = "black")
ax = plt.gca()
ax.add_patch(c1)
plt.axis("scaled")
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Figure1")

plt.scatter(x, y, marker=".", color = "green", label = "EN")
plt.show()
```
